# How to automatically clean and import a bunch of files into an SQL database  
This approach makes you manually upload large files because using \copy is much faster. 

## Optimized Version (Skip Large Files)

```python
from sqlalchemy import create_engine
import pandas as pd
import re
import os

# --- Database connection ---
engine = create_engine(
    "postgresql+psycopg2://postgres:f1sh3r@localhost:5432/Brazilian-E-Commerce"
)

# --- Folder path ---
folder_path = r"C:\Users\samww\Documents\Codecademy_projects\portfolio_projects\ChatGPT Project Brazilian E-Commerce Analysis\data archive"

# --- File size threshold (in MB) ---
# Adjust if needed, but ~5MB keeps runtime fast
SIZE_THRESHOLD_MB = 20

# --- Track skipped files ---
skipped_files = []

# --- Text cleaning function ---
def clean_text(x):
    return re.sub(r'[^\x20-\x7E]+', ' ', x) if isinstance(x, str) else x

# --- Loop through files ---
for file_name in os.listdir(folder_path):
    
    if file_name.endswith(".csv") and file_name.startswith("olist_"):
        
        file_path = os.path.join(folder_path, file_name)
        
        # --- Check file size ---
        file_size_mb = os.path.getsize(file_path) / (1024 * 1024)
        
        if file_size_mb > SIZE_THRESHOLD_MB:
            print(f"⏭ Skipping large file: {file_name} ({file_size_mb:.2f} MB)")
            skipped_files.append(file_name)
            continue
        
        print(f"\nProcessing: {file_name} ({file_size_mb:.2f} MB)")
        
        # --- Generate table name ---
        table_name = file_name.replace("olist_", "").replace("_dataset.csv", "")
        
        # --- Load data ---
        df = pd.read_csv(file_path, encoding='latin1', engine='python')
        
        # --- Clean text columns ---
        for col in df.select_dtypes(include=['object']).columns:
            df[col] = df[col].apply(clean_text)
        
        # --- Upload to PostgreSQL ---
        df.to_sql(
            table_name,
            engine,
            if_exists='replace',
            index=False,
            chunksize=5000,
            method='multi'
        )
        
        print(f"✅ Loaded into table: {table_name}")

# --- Print skipped files ---
print("\n⏭ Skipped files (too large):")
for f in skipped_files:
    print(f"- {f}")
```

### Manualy clean large files and upload them using \copy
olist_geolocation_dataset.csv

```python
import pandas as pd
import re

#Change the input and output paths to go to and from your skipped files. 
input_path = r"C:\Users\samww\Documents\Codecademy_projects\portfolio_projects\ChatGPT Project Brazilian E-Commerce Analysis\data archive\olist_geolocation_dataset.csv"
output_path = r"C:\Users\samww\Documents\Codecademy_projects\portfolio_projects\ChatGPT Project Brazilian E-Commerce Analysis\data archive\olist_geolocation_dataset_cleaned.csv"

# Read raw data
df = pd.read_csv(input_path, encoding='latin1', engine='python')

# Optional: remove any remaining non-printable characters
def clean_text(x):
    return re.sub(r'[^\x20-\x7E]+', ' ', x) if isinstance(x, str) else x #last part ensures that the function does not act on number, date, or geometry columns.
#re.sub(pattern, replacement, string)
#r'' (raw string) Prevents Python from interpreting escape characters
#\x20-\x7E → ASCII range: \x20 = space ( ) \x7E = ~. So this range includes all printable ASCII characters
#[ ... ] → character class Means: “match any ONE character inside this set”
#^ inside brackets → negation [^\x20-\x7E] Means: “match anything that is NOT a printable ASCII character”.

#Cleans all columns. 
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(clean_text)

# Save clean file
df.to_csv(output_path, index=False, encoding='utf-8')
```
